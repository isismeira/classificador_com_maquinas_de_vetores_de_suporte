# Classificação com Máquina de Vetores de Suporte (SVM)

Como visto anteriormente, o modelo feito com Árvores de Decisão ficou muito complexo devido à quantidade grande de atributos do dataset. Pode sofrer de um problema chamado maldição da dimensionalidade pois se especializa demais e pode generalizar demais

Muitas vezes com um subconjunto, é possível obter uma acurácia melhor do que com um conjunto completo. E com isso que se baseia a seleção de atributos

Dessa vez, será utilizado a Máquina de Vetor de Suporte que é um modelo que otimiza os vetores de suporte para classificação

Após isso, será utilizado uma seleção de atributos para mostrar uma métrica de importância de atributos

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.svm import SVC # para usar a Máquina de Vetor de Suporte
from sklearn.ensemble import ExtraTreesClassifier # para realizar a seleção de atributos
```

## Conhecendo o dataset

```python
dados = pd.read_csv('Credit.csv')
dados
```

```python
previsores = dados.iloc[:,0:20].values
classe = dados.iloc[:,20].values
```

## Transformação de atributos

```python
labelencoder = LabelEncoder()
previsores[:,0] = labelencoder.fit_transform(previsores[:,0])
previsores[:,2] = labelencoder.fit_transform(previsores[:,2])
previsores[:, 3] = labelencoder.fit_transform(previsores[:, 3])
previsores[:, 5] = labelencoder.fit_transform(previsores[:, 5])
previsores[:, 6] = labelencoder.fit_transform(previsores[:, 6])
previsores[:, 8] = labelencoder.fit_transform(previsores[:, 8])
previsores[:, 9] = labelencoder.fit_transform(previsores[:, 9])
previsores[:, 11] = labelencoder.fit_transform(previsores[:, 11])
previsores[:, 13] = labelencoder.fit_transform(previsores[:, 13])
previsores[:, 14] = labelencoder.fit_transform(previsores[:, 14])
previsores[:, 16] = labelencoder.fit_transform(previsores[:, 16])
previsores[:, 18] = labelencoder.fit_transform(previsores[:, 18])
previsores[:, 19] = labelencoder.fit_transform(previsores[:, 19])
```

## Divisão entre treino e teste

```python
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(previsores,
                                                                  classe,
                                                                  test_size = 0.3,
                                                                  random_state = 1)
```

## Treinamento e acurácia do modelo

```python
svm = SVC()
svm.fit(X_treinamento, y_treinamento)
```

```python
previsoes = svm.predict(X_teste)
previsoes
```

```python
taxa_acerto = accuracy_score(y_teste, previsoes)
taxa_acerto
```

Performance semelhante a do modelo de Naive Bayes, mas é possível tentar melhorar essa performance através da seleção de atributos

## Seleção de atributos

Verificando a importância de cada atributo

```python
forest = ExtraTreesClassifier()
forest.fit(X_treinamento, y_treinamento)
importancias = forest.feature_importances_
importancias
```

Criar um subconjunto com apenas os atributos de maior importância, que são os de índice 0, 1, 2, 4

```python
X_treinamento2 = X_treinamento[:,[0,1,2,3]]
X_teste2 = X_teste[:,[0,1,2,3]]
```

## Testando novamente a acurácia do modelo

Treinar e testar novamente o conjunto de testes mas agora com o subconjunto com os melhore atributos selecionados

```python
svm2 = SVC()
svm2.fit(X_treinamento2, y_treinamento)
previsoes2 = svm2.predict(X_teste2)
taxa_acerto = accuracy_score(y_teste, previsoes2)
taxa_acerto
```

Apesar da mudança ser insignificativa e talvez sujeita ao acaso já que se trabalha com amostragens, temos um modelo melhor, já que o modelo anterior possuia 20 variáveis independentes e agora, possui apenas 4 variáveis independentes
